# Polynomial Regression

## Load Dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Position_Salaries.csv')
print(df.head())
```

```python
X = df.iloc[:, 1:-1].values
print(X)
```

```python
y = df.iloc[:, -1].values
print(y)
```

## Training Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly, y)
```

## Visualize Linear regression

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, lr.predict(X), color='green')
plt.title('Simple linear regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

## Visuliza Polynomical regression

```python
plt.scatter(X, y, color='blue')
plt.plot(X, lr2.predict(X_poly), color='green')
plt.title('Polynomical regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

### Tip for more beautiful visualization

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='blue')
plt.plot(X_grid, lr2.predict(poly_reg.fit_transform(X_grid)), color='green')
plt.title('Polynomical regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
