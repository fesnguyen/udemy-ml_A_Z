```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing(as_frame=True)
df = housing.frame

df.to_csv('Data/house_price_train.csv')
```

```python
import matplotlib.pyplot as plt

for col in df.columns:
    plt.figure(figsize=(4, 2))
    plt.bar(df[col], df['HouseAge'])
    plt.xlabel(col)
    plt.ylabel('HouseAge')
    plt.title(f'Bar of {col}')
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the number of columns you want in the grid
n_cols = 3  # Adjust this number to change the number of charts per row
n_rows = (len(df.columns) + n_cols - 1) // n_cols  # Calculate the required number of rows

fig, axs = plt.subplots(n_rows, n_cols, figsize=(10, 15))  # Adjust the figsize accordingly
axs = axs.flatten()  # Flatten the array of axes to easily iterate over

# Iterate over the columns and plot each on a different subplot
for i, col in enumerate(df.columns.drop('MedHouseVal')):
    axs[i].scatter(df[col], df['MedHouseVal'], s=0.1)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('MedHouseVal')
    axs[i].set_title(f'Bar of {col}')

# Remove any unused subplots if there are more subplots than columns
for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
