# Naive Bayes

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
purchased_df = df[df['Purchased']==1]
not_purchased_df = df[df['Purchased']==0]
```

```python
plt.scatter(purchased_df['Age'].values, purchased_df['EstimatedSalary'].values, color='blue')
plt.scatter(not_purchased_df['Age'].values, not_purchased_df['EstimatedSalary'].values, color='red')
plt.show()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
print(X)
print(y)
```

## Split to train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train)
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

## Train Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

## Check model performance

```python
y_pred = classifier.predict(sc.transform(X_test).reshape(-1, 2))
print(np.concatenate((y_pred.reshape(-1, 1), y_test.reshape(-1, 1)), 1))
```

## Check model preformance using confusion_matric and accuracy_score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
